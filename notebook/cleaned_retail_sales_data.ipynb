**g. Further duplicate checking :**

**i.  checking all rows :**

```python
print(f"Total dropped duplicates are {df.duplicated().sum()}")

df = df.drop_duplicates().reset_index(drop=True)
```

**ii.  Checking duplicates based on `transation_id` :**

```python
suspicious = df[df.duplicated(subset='transaction_id', keep=False)]
print(f"Remaining suspicious duplicated rows: {len(suspicious)}")

for txn_id, group in suspicious.groupby('transaction_id'):
    mismatched_cols = [col for col in group.columns
                       if group[col].nunique(dropna=False) > 1]
    if mismatched_cols:  
        print(f"\ntransaction_id : {txn_id}")
        print(f"Mismatched cols: {mismatched_cols}")
        print(group[['transaction_id'] + mismatched_cols].to_string(index=False))
        print('-' * 60)
    else:
        print(f"\ntransaction_id : {txn_id} → Exact duplicate, need to be dropped")
```

**iii. Resolving `Quantity` and `product_name` Mismatches in Duplicate `transaction_id` Records Using Reverse Calculation :**

```python
df['product_name'] = df.groupby('transaction_id')['product_name'].transform(lambda x: x.ffill().bfill())

mask = df.duplicated(subset='transaction_id', keep=False)
suspicious = df[mask].sort_values('transaction_id')

only_qty_diff = suspicious.groupby('transaction_id').filter(
    lambda g: [c for c in g.columns if g[c].nunique(dropna=False) > 1] == ['quantity']
)

print(len(only_qty_diff))

only_qty_diff = only_qty_diff.copy()

only_qty_diff['actual_qty'] = ((only_qty_diff['total_sale'] + only_qty_diff['discount_amount']) / only_qty_diff['price_per_unit']).round(0).astype(int)

print(only_qty_diff[['transaction_id', 'quantity', 'actual_qty', 'price_per_unit', 'discount_amount', 'total_sale']].to_string(index=False))
```

**iv. Replace `quantity` by  `actual_qty` and dropped Duplicate :**

```python
df.loc[only_qty_diff.index, 'quantity'] = only_qty_diff['actual_qty']

print(f"Total dropped duplicates are  {df.duplicated().sum()}")

df = df.drop_duplicates().reset_index(drop=True)
```

**v.  Dropping Negative `price_per_unit` in Duplicate `transaction_id` Records:** 

```python
neg_price_idx = df[(df['transaction_id'] == 'Txn-01693') & (df['price_per_unit'] < 0)].index

df = df.drop(index=neg_price_idx).reset_index(drop=True)

print(f"Dropped {len(neg_price_idx)} row(s) with negative price_per_unit")
print(df[df['transaction_id'] == 'Txn-01693'][['transaction_id', 'price_per_unit']])
```

**vi. Dropping `Null` rows in Duplicate `transaction_id` Records:**

```python
for txn_id in ['Txn-01742', 'Txn-01983']:
    group = df[df['transaction_id'] == txn_id]
    
    drop_idx = group.isnull().sum(axis=1).idxmax()
    df = df.drop(index=drop_idx)

df = df.reset_index(drop=True)
print(f"Remaining rows: {len(df)}")
print(f"Remaining duplicate transaction_ids: {df.duplicated(subset='transaction_id').sum()}")
```

**ii. Handling missing values:**

**Step-1:** Drop Critical Missing Data by Removing Rows:

```python
df.dropna(subset=['product_name'], inplace = True)
df.reset_index(drop=True, inplace=True)
```

**Step-2:** Handling missing values of Numerical Data:

**i.** Checking negative numeric values:

```python
numeric_df = df.select_dtypes(include=['number'])
print((numeric_df < 0).sum())
```

**ii** Convert all negative value and NaN to zero:

```python
cols = ['age', 'quantity', 'price_per_unit', 'customer_rating']

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].mask((df[col] < 0) | (df[col].isna()), 0)
    
print(df[cols].eq(0).sum())
```

**iii.**  Handling Missing `age` column by using value of `age_group` column :

**Step-1:** Check age within age_group boundaries :

```python
def get_bounds_from_age_group(age_group):
    if pd.isna(age_group):
        return None
    age_group = str(age_group).strip()
    if age_group == '60+':
        return (61, 100)
    elif age_group == 'Under 18':
        return (1, 17)
    elif '-' in age_group:
        parts = age_group.split('-')
        return (int(parts[0]), int(parts[1]))
    return None

def is_out_of_bounds(row):
    if pd.isna(row['age']) or row['age'] == 0: 
        return False
    bounds = get_bounds_from_age_group(row['age_group'])
    if bounds is None:
        return False
    return not (bounds[0] <= row['age'] <= bounds[1])

df['age_invalid'] = df.apply(is_out_of_bounds, axis=1)

invalid_rows = df[df['age_invalid']]
print(f"Total out of boundary age rows: {len(invalid_rows)}")
print(invalid_rows[['transaction_id', 'age', 'age_group']].to_string(index=False))
```

**Step-2:** Covert false age into Zero:

```python
df.loc[df['age_invalid'], 'age'] = 0
df.drop(columns=['age_invalid'], inplace=True)
```

****Step-3:** Checking age column which providing zero velues:

```python
zero_age_rows = df[df['age'] == 0][['transaction_id', 'product_name', 'age', 'age_group']]

print(len(zero_age_rows))
print(zero_age_rows)
```

**Step-4:** Fills missing `age` values using the midpoint of the age_group range. For open-ended groups like `60+`, uses the midpoint between the group's lower bound and the dataset's actual maximum age.

```python
max_age = df['age'].max()
print(f'max value of age column :{max_age}')

def age_group_value(age_group):
    if '+' in age_group:
        lower_bound = int(age_group.replace('+', ''))
        return (lower_bound + max_age) // 2
    
    low, high = map(int, age_group.split('-'))
    return (low + high) // 2

df['age'] = df.apply(
           lambda row: age_group_value(row['age_group']) if row['age'] == 0 else row['age'], axis=1)

df['age'] = df['age'].round().astype(int)
```

**Step-5:** Further checking :

```python
print(f"Total null of age column: {df['age'].isnull().sum()}")
print(f"Total zero value of age column: {df['age'].eq(0).sum()}")
df.loc[zero_age_rows.index, ['transaction_id', 'product_name', 'age', 'age_group']].head()
```

**iv. Handling missing values of `quantity` and `price_per_unit` :**

**Step-1:** Checking missing values of `quantity` and `price_per_unit` :

```python
cols = ['quantity', 'price_per_unit']

zero_rows = df[df[cols].eq(0).any(axis=1)]

print(f"Total zero rows: {len(zero_rows)}")
print(zero_rows[['product_name', 'quantity', 'price_per_unit', 'total_sale', 'cogs', 'profit']])
```

**Step-2:** Fillup `quantity` column by using `total_sale` and `price_per_unit` columns :

```python
df['quantity'] = df.apply(
       lambda row: round(row['total_sale'] / row['price_per_unit'])
                   if row['quantity'] == 0 and row['price_per_unit'] != 0
                   else row['quantity'], axis=1
)
```

**Step-3: Fillup `price_per_unit` column by using `total_sale` and `quantity` columns :**

```python
df['price_per_unit'] = df.apply(
       lambda row: round(row['total_sale'] / row['quantity'], 2)
            if row['price_per_unit'] == 0 and row['quantity'] != 0
            else row['price_per_unit'], axis=1
)
```

**step-4: Further checking:**

```python
print(f"Total zero rows: {len(zero_rows)}")
print(f"Remaning zero value: \n {df[['quantity', 'price_per_unit']].eq(0).sum()}")
df.loc[zero_rows.index, ['product_name', 'quantity', 'price_per_unit', 'total_sale', 'cogs', 'profit']]
```

**vi. Handling missing values of Categorical Data:**

**i. Cross check of Calculated Columns:**

**Step-1:** Cross-check error calculation:

```python
TOLERANCE = 0.01

# --- Recalculate expected values ---
df['expected_discount_amount'] = (df['price_per_unit'] * df['quantity'] * df['discount_pct'] / 100).round(2)
df['expected_total_sale']      = ((df['price_per_unit'] * df['quantity']) - df['expected_discount_amount']).round(2)
df['expected_profit']          = (df['expected_total_sale'] - (df['cogs'] * df['quantity'])).round(2)
df['expected_profit_margin']   = ((df['expected_profit'] / df['expected_total_sale']) * 100).round(2)

# --- Flag mismatched rows ---
df['flag_discount_amount']  = ~np.isclose(df['discount_amount'],  df['expected_discount_amount'], atol=TOLERANCE)
df['flag_total_sale']       = ~np.isclose(df['total_sale'],       df['expected_total_sale'],      atol=TOLERANCE)
df['flag_profit']           = ~np.isclose(df['profit'],           df['expected_profit'],           atol=TOLERANCE)
df['flag_profit_margin']    = ~np.isclose(df['profit_margin_pct'],df['expected_profit_margin'],    atol=TOLERANCE)

# --- Any row that has at least one mismatch ---
any_mismatch = df['flag_discount_amount'] | df['flag_total_sale'] | df['flag_profit'] | df['flag_profit_margin']

# --- Print mismatch summary ---
print("=== Mismatch Count ===")
print(f"discount_amount   : {df['flag_discount_amount'].sum()} rows")
print(f"total_sale        : {df['flag_total_sale'].sum()} rows")
print(f"profit            : {df['flag_profit'].sum()} rows")
print(f"profit_margin_pct : {df['flag_profit_margin'].sum()} rows")
print(f"\nTotal rows with at least one mismatch: {any_mismatch.sum()}")

# --- Print mismatched rows: original vs expected side by side ---
print("\n=== Mismatched Rows (Original vs Expected) ===")
mismatch_view = df.loc[any_mismatch, [
    'transaction_id',
    'quantity', 'price_per_unit', 'cogs', 'discount_pct',
    'discount_amount',   'expected_discount_amount',
    'total_sale',        'expected_total_sale',
    'profit',            'expected_profit',
    'profit_margin_pct', 'expected_profit_margin'
]]
```

**Step-2:** Replace all mismatched values with recalculated ones:

```python
# ── Replace all mismatched values with recalculated ones ──────────
df.loc[any_mismatch, 'discount_amount']   = df.loc[any_mismatch, 'expected_discount_amount']
df.loc[any_mismatch, 'total_sale']        = df.loc[any_mismatch, 'expected_total_sale']
df.loc[any_mismatch, 'profit']            = df.loc[any_mismatch, 'expected_profit']
df.loc[any_mismatch, 'profit_margin_pct'] = df.loc[any_mismatch, 'expected_profit_margin']

print(f"Replaced {any_mismatch.sum()} mismatched rows with recalculated values")

# ── Drop helper columns ───────────────────────────────────────────
df.drop(columns=[
    'expected_discount_amount', 'expected_total_sale',
    'expected_profit', 'expected_profit_margin',
    'flag_discount_amount', 'flag_total_sale',
    'flag_profit', 'flag_profit_margin'
], inplace=True)

print(f"Remaining rows : {len(df)}")
```
